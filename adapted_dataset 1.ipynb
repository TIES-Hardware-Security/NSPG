```python
import pandas as pd

# Function to read a JSON file and convert it into a DataFrame
def json_to_dataframe(file_path):
    # Load the JSON data from the file
    data = pd.read_json(file_path)
    
    # Convert the JSON data into a DataFrame
    df = pd.DataFrame(data)

    return df

# Example usage
file_path = './DATA/casedataset-!if.json'
df = json_to_dataframe(file_path)
df = df.sample(frac=1).reset_index(drop=True)
# trim the data to 1000 rows
df = df[:1000]
```

```python
# append the new data to the existing data

file_path = './DATA/casedataset-if.json'
df2 = json_to_dataframe(file_path)
df2 = df2.sample(frac=1).reset_index(drop=True)
# trim the data to 1000 rows
df2 = df2[:2000]
```

```python
file_path = './DATA/ifdataset.json'
df3 = json_to_dataframe(file_path)
df3 = df3.sample(frac=1).reset_index(drop=True)
# trim the data to 1000 rows
df3 = df3[:2000]
```

```python
file_path = './DATA/ifdataset-else.json'
df4 = json_to_dataframe(file_path)
df4 = df4.sample(frac=1).reset_index(drop=True)
# trim the data to 1000 rows
df4 = df4[:2000]
```

```python
df = df._append(df2, ignore_index=True)
df = df._append(df3, ignore_index=True)
df = df._append(df4, ignore_index=True)
# shuffle the rows
df = df.sample(frac=1).reset_index(drop=True)
```

```python
str = '''You are an AI programming assistant, utilizing the Deepseek Coder model, developed by Deepseek Company, and you only answer questions related to computer science. For politically sensitive questions, security and privacy issues, and other non-computer science questions, you will refuse to answer.
### Instruction:
{prompt}
### Response:
'''
output = []

for index, row in df.iterrows():
    Text = "You are an AI programming assistant, utilizing the Deepseek Coder model, developed by Deepseek Company, and you only answer questions related to computer science. For politically sensitive questions, security and privacy issues, and other non-computer science questions, you will refuse to answer. \n ### Instruction: "
    Text += "Generate a list of systemverilog assertions from the following code \n" + row['Code']
    Text += "\n ### Response: \n Here is the requested property \n\n " + row['Assertion']
    Text += "\n[EOT]"
    output.append({'text': Text})

import json

# Write the formatted JSON to a file
with open('./DATA/Adapted_dataset.json', 'w') as f:
    f.write(json.dumps(output, indent=4))
```
